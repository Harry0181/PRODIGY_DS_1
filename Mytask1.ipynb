```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
df = pd.read_excel('/content/d1.xlsx')
df.set_index('Country Name', inplace=True)
```

```python
for country in df.index:
    plt.figure(figsize=(6, 4))  # Adjust the size as needed
    plt.hist(df.columns, bins=len(df.columns), weights=df.loc[country], edgecolor='black')
    plt.gca().yaxis.set_major_formatter(ticker.ScalarFormatter())
    plt.title(f'Population Histogram for {country}')
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.xticks(rotation=45)
    plt.tight_layout()  # Adjust layout to fit labels
    plt.show()
```
